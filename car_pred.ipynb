```python
import pandas as pd

file_name = 'car_details_chennai.json'
df = pd.read_json(file_name)
print(df)

import pandas as pd

csv_file_name = 'car_details_chennai.csv'
df.to_csv(csv_file_name, index=False)
pd.set_option('display.max_columns', None)
chennai = pd.read_csv(csv_file_name)

columns_to_drop = [
    "Year of Manufacture", "Engine", "Power", "Seating Capacity", "Drive Type", 
    "Mileage","Fuel", "Top Speed", "Ground Clearance", "No. of Airbags", "Length", 
    "Range", "Battery Capacity", "Boot Space", "Global NCAP Safety Rating", 
    "Charging Time DC", "Charging Time AC",
]


chennai.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
chennai.isnull().sum()
```

```python
chennai['title'].unique()
```

```python
chennai['title'] = chennai['title'].astype(str)
```

```python
chennai[['Brand', 'Model']] = chennai['title'].str.extract(r'^(?:\d{4}\s+)?(\S+)\s+(.+)$')
```

```python
chennai['Brand'].fillna('Unknown', inplace=True)
chennai['Model'].fillna('Unknown', inplace=True)
```

```python
chennai.drop(columns=['title'], inplace=True)
```

```python
chennai
```

```python
chennai.columns= ["Price","Location","Registration_Year","Insurance","Fuel_Type","Seats","Kms_Driven","RTO","Ownership","Engine_Displacement","Transmission","Brand","Model"]

chennai = chennai[sorted(chennai.columns)]
```

```python
chennai
```

```python
chennai.isnull().sum()
```

```python
col = ["Engine_Displacement","Ownership","RTO","Registration_Year"]

for i in col:
    chennai[i].fillna(method = 'bfill', inplace=True)
```

```python
chennai.isnull().sum()
```

```python
chennai.duplicated().sum()
```

```python
chennai.drop_duplicates(keep ='last')
```

```python
chennai.duplicated().sum()
```

```python
duplicates = chennai[chennai.duplicated()]
print(duplicates)
```

```python
chennai.drop_duplicates(keep=False, inplace=True)
```

```python
chennai.duplicated().sum()
```

```python
chennai.to_csv('car_details_chennai.csv', index=False)
```

```python
import pandas as pd

# List of multi-word brands
multi_word_brands = {
    "Land": "Land Rover",
    "Mini": "Mini Cooper",
    "Mercedes-Benz": "Mercedes Benz"
}

# Function to correct the brand
def correct_brand_model(row):
    brand, model = row['Brand'], row['Model']
    if brand in multi_word_brands:
        brand = multi_word_brands[brand]  # Correct brand
        model = " ".join(model.split()[1:])  # Remove the extra brand word from the model
    return pd.Series([brand, model])

# Apply the correction
car[['Brand', 'Model']] = car.apply(correct_brand_model, axis=1)

```

```python

import re


# Define patterns to remove (engine size, transmission, fuel type, etc.)
remove_patterns = [
    r'\b\d+\.\d+\b',  # Engine size like 1.2, 3.0, 520d
    r'\b\d+l\b',  # Engine size with 'l' (e.g., 3.0 l, 2.5L)
    r'\b\d+[dD]\b',  # Engine size with 'd' (e.g., 520d, 350d)
    r'\b(Petrol|Diesel|CNG|EV|Hybrid|BSIV|BSVI|LPG)\b',  # Fuel types and BS norms
]

# Function to clean the Model column
def clean_model_name(model_name):
    if pd.isna(model_name):  # Handle missing values
        return ""
    cleaned_model = model_name.strip()
    for pattern in remove_patterns:
        cleaned_model = re.sub(pattern, '', cleaned_model, flags=re.IGNORECASE).strip()
    return cleaned_model

# Apply cleaning function to the Model column
car['Model'] = car['Model'].apply(clean_model_name)

```

```python
import mlflow
import mlflow.sklearn
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mlflow.set_experiment("Car_Price_Prediction")

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "XGBoost": xgb.XGBRegressor(),
    "SVR": SVR()
}

for model_name, model in models.items():
    with mlflow.start_run(run_name=model_name):
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

        # Calculate Metrics
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Log parameters and metrics
        mlflow.log_param("Model", model_name)
        mlflow.log_metric("Mean Squared Error", mse)
        mlflow.log_metric("Mean Absolute Error", mae)
        mlflow.log_metric("R² Score", r2)

        # Log the model
        mlflow.sklearn.log_model(model, model_name)

        print(f"{model_name}:")
        print(f"  Mean Squared Error: {mse:.2f}")
        print(f"  Mean Absolute Error: {mae:.2f}")
        print(f"  R² Score: {r2:.2f}\n")

print("All models trained and logged in MLflow.")
```
